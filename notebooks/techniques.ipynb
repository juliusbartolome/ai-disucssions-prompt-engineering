# Prompting Techniques

````python
import os
import requests

API_KEY = os.environ.get("OPENAI_API_KEY")
MODEL = os.environ.get("OPENAI_MODEL")

if not API_KEY or not MODEL:
    raise Exception("Environment variables OPENAI_API_KEY and OPENAI_MODEL must be set.")

API_URL = f"https://api.openai.com/v1/chat/completions"
print(MODEL)

headers = {
  "Authorization": f"Bearer {API_KEY}",
  "Content-Type": "application/json"
}

def query(prompt):
  request_json = {
    "model": MODEL,
    "messages": [
      {
        "role": "user",
        "content": prompt
      }
    ]
  }

  response = requests.post(API_URL, headers=headers, json=request_json)
  response_json = response.json()
  # print(response_json)
  return response_json["choices"][0]["message"]["content"]
````

## 1. Zero-Shot

### Description:

Imagine whispering an exact wish to a powerful language genie. That's zero-shot prompting! You guide large language models with simple instructions (prompts) to achieve anything from poetic verses to informative summaries, even without training them specifically.

### Best Use Cases:

* **Creative Spark:** Fuel your imagination with diverse tasks like writing poems, scripts, or musical pieces. No coding required, just craft the perfect prompt and watch the model ignite your ideas.
* **Knowledge Seeker:** Get instant insights from articles, documents, or research papers. Summarize key points, extract specific information, or answer your burning questions – all with a well-constructed prompt.
* **Productivity Booster:** Automate repetitive tasks like text summarization, translation, or email writing. Free up your time for high-level thinking and strategic work.

### Examples:

- **Prompt:** 
  Craft a suspenseful short story opening set in a moonlit forest.

- **Response:**

#### 2. **Information Alchemist:**

- **Prompt:** 
  Briefly summarize the main arguments of the scientific paper "Climate Change and its Impact on Ocean Ecosystems."

- **Response:** 
  The study highlights the detrimental effects of rising temperatures and ocean acidification on marine life. Coral reefs and delicate ecosystems face increased stress, leading to biodiversity loss and potential collapse of vital food webs. Urgent action is needed to mitigate climate change and protect our oceans.

#### 3. **Test Yourself:**

- **Prompt:** 
  Generate a catchy slogan for a brand of sustainable, energy-efficient home appliances.


Feel free to unleash your creativity and see what amazing things you can conjure up with the magic of zero-shot prompting! Remember, the clearer your instructions, the more impressive the results. Happy prompting!


````python
text_prompt = "Generate a catchy slogan for a brand of sustainable, energy-efficient home appliances."

output_text = query(text_prompt)

print(output_text)
````

## 3. Chain-of-Thought

### Description

Forget brute-force training! Chain-of-thought prompting empowers you to guide LLMs with insightful breadcrumbs, leading them to the desired outcome through a series of logical steps. Think of it as crafting a map, guiding your AI explorer to a specific destination instead of letting it wander aimlessly. This technique unlocks deeper reasoning capabilities, enabling LLMs to tackle complex tasks beyond simple pattern recognition.

### Best Use Cases

* **Step-by-step Problem Solving:** Break down intricate challenges into smaller, manageable steps, empowering the LLM to reason through each stage and arrive at the final solution. Ideal for complex math problems, scientific reasoning, or crafting detailed narratives.
* **Explainable AI:** Gain transparency into the LLM's thought process by observing the chain of reasoning it takes to reach a conclusion. This allows for debugging, error analysis, and building trust in the LLM's outputs.
* **Knowledge Fusion:** Combine your expertise with the LLM's vast knowledge base. Provide a series of examples demonstrating your desired logic, and let the model use those steps to analyze new data or generate creative outputs aligned with your specific insights.

### Examples:

**1. Medical Sleuth:**

- **Prompt:**

  Patient A with symptoms X, Y, and Z was diagnosed with disease D.
  
  Patient B with similar symptoms, but also symptom W, was also diagnosed with disease D.
  
  Patient C presents with symptoms X, Y, Z, and W.
  
  Based on the reasoning applied to previous cases, what is the most likely diagnosis for Patient C?

- **Response:**

  Following the established chain of thought, Patient C exhibits all known symptoms of disease D and shares the additional symptom W observed in one confirmed case. Therefore, the most likely diagnosis for Patient C is also disease D. Further tests are recommended for definitive confirmation.

**2. Code Composer:**

- **Prompt:**

  Function A takes input X and returns result Y.

  Function B takes input Y and performs computation Z.

  Function C needs to combine the functionalities of A and B, taking input X and performing both computations.

  Write the code for Function C, leveraging the logic and functionalities of A and B in sequence.

- **Response:**

  ```python
  def function_C(x):
      y = function_A(x)
      result = function_B(y)
      return result
  ```

**3. Test Yourself:**

- **Prompt:**

  A news article reports a sudden increase in a specific bird population.

  Experts attribute the rise to a recent change in the local environment.

  Write a short social media post raising awareness about the environmental change and its potential impact, using facts and logic established in the previous steps.

**Remember:**

The more detailed and logical your chains of thought, the more accurate and insightful the LLM's responses will be. Embrace the power of step-by-step reasoning and unlock the LLM's full potential with chain-of-thought prompting!

This format provides a comprehensive overview of chain-of-thought prompting, highlighting its benefits and showcasing its application through clear examples. Feel free to adapt and modify this template to explore various tasks and scenarios based on your specific interests and needs!


````python
text_prompt = """
A news article reports a sudden increase in a specific bird population.
Experts attribute the rise to a recent change in the local environment.
Write a short social media post raising awareness about the environmental change and its potential impact, using facts and logic established in the previous steps.
"""

output_text = query(text_prompt)
print(output_text)
````

By gently nudging the LLM towards consistency within its own responses, you unlock remarkable accuracy and reasoning across diverse tasks. Think of it as a mindful dialogue, shaping the model's internal logic step-by-step.

### Best Use Cases:

* **Concise Perfection:** Achieve elegant and accurate results, without the need for lengthy chains of thought. Ideal for numerical computations, logical puzzles, or tasks where internal consistency is paramount.
* **Debugging Delight:** Identify inconsistencies and errors by observing the LLM's attempts to reconcile its reasoning. Perfect for uncovering flaws in logic or ensuring factual accuracy.
* **Creative Exploration:** Encourage innovative outputs by prompting internal exploration and self-evaluation within the LLM's vast knowledge base. Ideal for generating novel ideas, writing coherent narratives, or crafting unique artistic expressions.

### Examples:

**1. Number Ninja:**

- **Prompt:**

  1. Start with the number 5.
  2. Double it to get 10.
  3. Add 3 to get 13.
  4. What is the final number?

- **Response:**

  - (Through internal consistency checks, the LLM recognizes double 5 leads to 10, not 13. Therefore, the final number is 10.)

**2. Story Weaver:**

- **Prompt:**

  1. Once upon a time, a brave knight…
  2. The knight embarked on a quest to find…
  3. But a mysterious sorceress…
  4. Who will triumph in the end?

- **Response:**

  - (The LLM weaves a consistent narrative with a knight facing a sorceress, ensuring logical coherence throughout the story.)

**3. Test Yourself:**

- **Prompt:**

  1. Imagine a world where gravity works in reverse.
  2. How would everyday activities like walking or eating be affected?
  3. Write a short poem capturing the essence of this upside-down world.

**Remember:**

The key to self-consistency prompting lies in providing subtle hints and encouraging internal reasoning. Trust the LLM's inherent logic and allow it to navigate its own knowledge pathways. Be surprised by the elegant accuracy and creative potential that unfolds!

By embracing this gentle guidance, you unlock the LLM's true potential to reason, innovate, and deliver remarkable results with a captivating touch of self-aware consistency.

## 6. Prompt Chaining

### Description

Forget one-shot wonders! Prompt chaining transforms complex tasks into bite-sized steps, building on each LLM response to reach the final goal. Think of it as a relay race, where you pass the baton of information, refining the output with each prompt until you reach the finish line. This technique unlocks the LLM's potential for intricate tasks, breaking them down into manageable chunks for enhanced accuracy and control.

### Best Use Cases:

* **Conquering Complexity:** Tackle multi-faceted challenges by providing the LLM with a series of focused prompts, guiding it through each subtask to achieve the overall objective. Ideal for advanced data analysis, creative storytelling, or intricate problem-solving.
* **Iterative Exploration:** Refine your desired outcome with each chained prompt. Use the LLM's initial responses as stepping stones, providing feedback and guidance to shape the final result. Perfect for crafting personalized creative outputs, adjusting technical specifications, or honing research findings.
* **Error Debugging:** Isolate and pinpoint errors by breaking down the task into smaller chains. Analyze each LLM response in context, identifying missteps and refining subsequent prompts to ensure accuracy. A powerful tool for scientific simulations, technical documentation, or ensuring factual correctness.

### Examples:

**1. Master Chef Chaining:**

- **Chain 1:** Describe a delicious and healthy dessert recipe.
- **Response:** A light and refreshing panna cotta with mango purée and coconut milk.
- **Chain 2:** What can I add to the panna cotta for a burst of protein?
- **Response:** Chia seeds or hemp seeds would be great options, adding richness and a satisfying texture.
- **Final Recipe:** A protein-packed panna cotta with chia seeds, topped with a vibrant mango and coconut milk purée.

**2. Mystery Novel Maze:**

- **Chain 1:** Introduce the detective protagonist and their unique investigative style.
- **Response:** Amelia Vance, a sharp-witted detective with a passion for cryptic puzzles and a talent for reading people.
- **Chain 2:** Describe the initial crime scene in a secluded mansion.
- **Response:** An opulent library, the fireplace cold, a priceless painting slashed upon the canvas.
- **Chain 3:** Who are the main suspects and their motives?
- **Response:** The eccentric art collector, the jealous rival painter, and the enigmatic housekeeper with a hidden past.

**3. Test Yourself:**

- **Chain 1:** Imagine a sustainable city powered by renewable energy sources.
- **Chain 2:** Describe the types of renewable energy sources used and their specific applications.
- **Chain 3:** Identify potential challenges in implementing this technology on a large scale.

**Remember:**

The key to successful prompt chaining is clarity and focus. Provide precise prompts for each subtask, building upon the LLM's responses with insightful feedback. With each chain, you refine the model's understanding, leading to a final outcome that surpasses the sum of its parts.

Embrace the power of prompt chaining, and watch your LLMs tackle complex challenges with newfound agility and accuracy!

````python
text_prompt = "Imagine a sustainable city powered by renewable energy sources."
output_text = query(text_prompt)
print("Prompt:")
print(text_prompt)
print("\nResponse:")
print(output_text)

print("\n-------------------------------------------------------------------\n")

text_prompt = "Describe the types of renewable energy sources used and their specific applications."
output_text = query(text_prompt)
print("Prompt:")
print(text_prompt)
print("\nResponse:")
print(output_text)

print("\n-------------------------------------------------------------------\n")

text_prompt = "Identify potential challenges in implementing this technology on a large scale."
output_text = query(text_prompt)
print("Prompt:")
print(text_prompt)
print("\nResponse:")
print(output_text)
````
